```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.linear_model import *
from sklearn.metrics import mean_squared_error
```

```python
wine2.head()
```

```python
wine3 = wine2[['country','price','province','variety','winery', 'points']]
```

```python
wine4 = wine3.sample(n=20000, random_state=42)
```

```python
wine4.info()
```

```python
wine4 = pd.get_dummies(wine4)
```

```python
wine4
```

```python
Y= wine4['points']
X = wine4.drop(['points'], axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=0.3, random_state=123)
```

```python
lr = LinearRegression(n_jobs=5000)
lo = LogisticRegression()
```

```python
lr.fit(X_train, Y_train)
lo.fit(X_train, Y_train)
```

```python
lr_Y_predict = lr.predict(X_test)
lo_Y_predict = lr.predict(X_test)
```

```python
lr_mse = mean_squared_error(Y_test, lr_Y_predict, squared=True)
lr_rmse = mean_squared_error(Y_test, lr_Y_predict, squared=False)

lo_mse = mean_squared_error(Y_test, lo_Y_predict, squared=True)
lo_rmse = mean_squared_error(Y_test, lo_Y_predict, squared=False)

print(f"Linear Regression   MSE: {lr_mse}  RMSE: {lr_rmse}\nLogistic Regression   MSE: {lo_mse}  RMSE: {lo_rmse}")
```
